```python
from Bio import pairwise2
from Bio.pairwise2 import format_alignment
from Bio import Entrez
from Bio import SeqIO
from colorama import Fore

#Fetch protein coding sequence of gene 1 from Entrez Database
def fetch_gene1(id):
    Entrez.email = "aanya.kamath@gmail.com"
    handle = Entrez.efetch(db="nucleotide", id="NM_007319.1", rettype="gb", retmode="text")
    record = SeqIO.read(handle, "genbank")
    handle.close()
    return record

#Fetch protein coding sequence of gene 2 from Entrez Database
def fetch_gene2(id):
    Entrez.email = "aanya.kamath@gmail.com"
    handle = Entrez.efetch(db="nucleotide", id="NM_007318.3", rettype="gb", retmode="text")
    record=SeqIO.read(handle, "genbank")
    handle.close()
    return record

#Defining the initial region of gene 1
start_position1 = 1
end_position1 = 70

sequence_record1 = fetch_gene1(id)
gene1_sequence = sequence_record1.seq

#Printing the description of gene 1
print(f"Gene 1 Accession Number: {sequence_record1.id}")
print(f"Description: {sequence_record1.description}")

region_1 = gene1_sequence[start_position1-1:end_position1]

#Defining the initial region of gene 2
start_position2 = 1
end_position2 = 70 

sequence_record2 = fetch_gene2(id)
gene2_sequence = sequence_record2.seq

#Printing the description of gene 2
print(f"Gene 2 Accession Number: {sequence_record2.id}")
print(f"Description: {sequence_record2.description}")

while end_position2 <= 73:
    region_2 = gene2_sequence[start_position2-1:end_position2]
    print(f"Gene 1 Sequence: {region_1}")
    print(f"Gene 2 Sequence: {region_2}")
    #Running the pairwise2 global alignment function to produce a similarity score between the defined regions of gene 1 and gene 2
    for a in pairwise2.align.globalms(region_1, region_2, 2, -1, -.5, -.1):
        print(format_alignment(*a))
    print(f"{Fore.RED}Comparison Finished!{Fore.BLACK}")
    start_position2 += 1
    end_position2 += 1
    #Adjusting the boundaries of the region of gene 2 by one nucleotide
```
